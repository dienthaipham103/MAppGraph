## AppScanner

Cut source files into traffic segments of 50 minutes 

```python
# get time of each file
source_folder = os.path.join(root_path, 'sources')
data_folder = os.path.join(root_path, '5_3/appscanner_75/full_data')
segment_length = 75 # minutes

for app in os.listdir(source_folder):
# for app in apps_8:
  print('Processing %s ...' % app)
  app_folder = os.path.join(source_folder, app)
  new_app_folder = os.path.join(data_folder, app)
  os.mkdir(new_app_folder)

  for filename in os.listdir(app_folder):
    t = float(filename[:-5].split('_')[-1])
    if t >= 50:
      index = 1

      file_path = os.path.join(app_folder, filename)
      df = pd.read_csv(file_path, index_col=0)
      base = df['time'].iloc[0]
      end = df['time'].iloc[-1]
      
      while ((index - 1)*segment_length + segment_length)*60 + base < end:
        start_time = base + (index-1)*segment_length*60
        end_time = start_time + segment_length*60
        df_ = df[(df['time'] >= start_time) & (df['time'] <= end_time)].reset_index(drop=True)

        # save to csv file
        new_filename = "_".join(filename.split('_')[:-2]) + '_' + filename.split('_')[-2] + '_' + str(index) + '.csv'
        new_file_path = os.path.join(new_app_folder, new_filename)
        df_.to_csv(new_file_path, index=True)

        index += 1
  
  print('.....................................................................')
```
