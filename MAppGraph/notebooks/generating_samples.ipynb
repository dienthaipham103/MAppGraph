## Cut into samples of 5 minutes (overlap 2 minutes)

```python
"""
cut all files (sources/app) into samples of 5 minutes with overlap of 2 minutes (samples/app, ex. baohay24h_1_1.csv) 
"""

sample_length = 1 # minutes
overlap_length = 0 # minutes

sources_folder = os.path.join(root_path, 'sources')
samples_folder = os.path.join(root_path, '1_0/samples')


# for app in os.listdir(sources_folder):
# for app in apps[90:]:
for app in apps_8:
  print('Processing %s ...' % app)
  app_folder = os.path.join(sources_folder, app)

  # create new app folder
  new_app_folder = os.path.join(samples_folder, app)
  if not os.path.exists(new_app_folder):
    os.mkdir(new_app_folder)

    for filename in os.listdir(app_folder):
      print(filename)
      index = 1

      file_path = os.path.join(app_folder, filename)
      df = pd.read_csv(file_path, index_col=0)
      base = df['time'].iloc[0]
      end = df['time'].iloc[-1]
      
      while ((index - 1)*(sample_length - overlap_length) + sample_length)*60 + base < end:
        start_time = base + (index-1)*(sample_length - overlap_length)*60
        end_time = start_time + sample_length*60
        df_ = df[(df['time'] >= start_time) & (df['time'] <= end_time)].reset_index(drop=True)

        # save to csv file
        new_filename = "_".join(filename.split('_')[:-2]) + '_' + filename.split('_')[-2] + '_' + str(index) + '.csv'
        new_file_path = os.path.join(new_app_folder, new_filename)
        df_.to_csv(new_file_path, index=True)

        index += 1
        # if index % 5 == 0:
        #   print(index)
    
    print('...................................................')
```

## Count samples for each class

```python
"""
count number of sample points for each app and save into json file
"""

samples_folder = os.path.join(root_path, '1_0/samples')
app_num = 0
samples_num = dict()
meta_info = dict()

for app in os.listdir(samples_folder):
  app_num += 1
  count = 0

  for _ in os.listdir(os.path.join(samples_folder, app)):
    count += 1
  samples_num[app] = count

meta_info['apps_number'] = app_num
meta_info['samples'] = samples_num

print(samples_num)
print(meta_info)

with open('/content/drive/My Drive/AI PROJECTS/GNN/dataset/1_0/samples.json', 'w') as f:
    json.dump(meta_info, f)
```
