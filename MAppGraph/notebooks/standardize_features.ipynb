## Standardize features



```python
# load normalize info
info_path = '/content/drive/My Drive/AI PROJECTS/GNN/dataset/5_3/standardization_N10.json'
with open(info_path, 'r') as f:
    mean_std_dic = json.load(f)
```

```python
# norm train graphs
for app in classes:
  f1_path = os.path.join('/content/drive/My Drive/AI PROJECTS/GNN/dataset/5_3/train_graphs/N10/t10', app, 'features.csv')
  w1_path = os.path.join('/content/drive/My Drive/AI PROJECTS/GNN/dataset/5_3/train_graphs/N10/t10', app, 'weights.csv')

  f_df = pd.read_csv(f1_path, index_col=0)
  # normalize the features dataframe
  for feature in mean_std_dic.keys():
    m, std = mean_std_dic[feature][0], mean_std_dic[feature][1]
    f_df[feature] = (f_df[feature] - m)/std
  
  # normalize ip feature
  f_df['ip1'] = f_df['ip1']/255
  f_df['ip2'] = f_df['ip2']/255
  f_df['ip3'] = f_df['ip3']/255
  f_df['ip4'] = f_df['ip4']/255

  # create app folder
  app_f = os.path.join('/content/drive/My Drive/AI PROJECTS/GNN/dataset/5_3/stand_train_graphs/N10/t10', app)
  os.mkdir(app_f)
  f2_path = os.path.join(app_f, 'features.csv')
  w2_path = os.path.join(app_f, 'weights.csv')

  # save features df
  f_df.to_csv(f2_path)
  # copy weights
  shutil.copy(w1_path, w2_path)

  print(app)
```
