## Distribution of weights

```python
def analyse_weight(path, T, t, norm):
  df = pd.DataFrame([], columns=['source', 'target', 'weight', 'graph_id'])

  for cls in classes:
    csv_path = os.path.join(path, cls, 'weights.csv')
    df_ = pd.read_csv(csv_path, index_col=0)
    df = pd.concat([df, df_], axis=0)
  
  df = df.reset_index(drop=True)

  # plot thd distribution
  fig, ax = plt.subplots(figsize=(10, 8), ncols=1, nrows=1)
  sns.distplot(df['weight'])
  ax.set_xlabel('w')
  ax.set_title('Distribution of w')
  plt.show()

  # statistic
  total = df.shape[0]
  if norm == False:
    n = int(T*60/t) + 1
    for i in range(n):
      sub_df = df[df['weight'] == i]
      count = sub_df.shape[0]
      per = round(count/total*100, 2)
      print('w = {}: {} %'.format(i, per))
  else:
    sub_df = df[df['weight'] > 0]
    count = sub_df.shape[0]
    per = round(count/total*100, 2)
    print('w = 0: {} %'.format(100 - per))
    print('w > 0: {} %'.format(per))
```

```python
analyse_weight('/content/drive/My Drive/AI PROJECTS/GNN/dataset/5_3/train_graphs/N30/t_10/', 5, 10, norm=False)
```
