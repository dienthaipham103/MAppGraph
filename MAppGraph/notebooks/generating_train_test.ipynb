## Generate training and testing randomly

Note: Testing data for all 3 models, but training dataset for GNN, Flowprint are different from Appscanner

```python
"""
input: list of filenames (samples) and the train size
output: list of filenames for training, list of filenames of testing
"""

def get_train_test(files_list, train_size):
  train_idx = random.sample(range(len(files_list)), int(len(files_list)*train_size))
  train_files = [files_list[i] for i in train_idx]

  test_files = list(set(files_list) - set(train_files))

  return (train_files, test_files)
```

```python
samples_folder = os.path.join(root_path, '1_0/samples')
train_size = 0.8
train_test_info = dict()

for app in os.listdir(samples_folder):
  app_folder = os.path.join(samples_folder, app)
  files_list = os.listdir(app_folder)

  train_test_info[app] = get_train_test(files_list, train_size)

"""
save train_test_info as json file
{
  app: [train_files_list], [test_files_list]
}
"""
with open('/content/drive/My Drive/AI PROJECTS/GNN/dataset/1_0/train_test_info.json', 'w') as f:
    json.dump(train_test_info, f)
```

```python
"""
Copy samples into training folder and testing folder
"""
train_folder = os.path.join(root_path, '1_0/train')
test_folder = os.path.join(root_path, '1_0/test')

with open('/content/drive/My Drive/AI PROJECTS/GNN/dataset/1_0/train_test_info.json', 'r') as f:
  train_test_info = json.load(f)
```

```python
########################### copy train files #################################
count1 = 0
for app in train_test_info.keys():
  count1 += 1
  print(count1)

  # create folder for app
  app_folder1 = os.path.join('/content/drive/My Drive/AI PROJECTS/GNN/dataset/1_0/samples', app)
  app_folder2 = os.path.join(train_folder, app)

  if not os.path.exists(app_folder2):
    os.mkdir(app_folder2)

    filenames = train_test_info[app][0]
    train_num = len(filenames)
    files_num = len(os.listdir(app_folder2))

    for filename in filenames:
      path1 = os.path.join(app_folder1, filename)
      path2 = os.path.join(app_folder2, filename)
      if not os.path.exists(path2):
        shutil.copy(path1, path2)
```
