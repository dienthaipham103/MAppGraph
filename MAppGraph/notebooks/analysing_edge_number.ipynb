```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
import json
import seaborn as sns
import matplotlib.pyplot as plt
```

----------------------------------------------------------------------------

```python
# Path to the graphs
root_path = '/content/drive/My Drive/AI PROJECTS/GNN/dataset/5_3/train_graphs/N30' 
```

```python
def plot_distribution(N_path, t):
  t_path = os.path.join(N_path, 't%d'%t)
  edge_nums = []

  # gen the number of edges of all graphs
  for filename in os.listdir(t_path):
    w_path = os.path.join(t_path, filename, 'weights.csv')
    df = pd.read_csv(w_path, index_col=0)
    graph_num = df['graph_id'].iloc[-1]
    for i in range(1, graph_num + 1):
      graph_df = df[(df['graph_id'] == i) & (df['weight'] > 0.0)]
      edge_num = graph_df.shape[0]
      edge_nums.append(edge_num)
  
  # plot the distribution
  fig, ax = plt.subplots(figsize=(10, 8), ncols=1, nrows=1)
  sns.histplot(edge_nums)
  ax.set_xlabel('Number of edges')
  ax.set_title('5 MINUTES TRAFFIC, N=30, t=%ds'%t)

  plt.show()
```

```python
plot_distribution(root_path, 10)
```

```python
plot_distribution(root_path, 5)
```

```python
plot_distribution(root_path, 1)
```
