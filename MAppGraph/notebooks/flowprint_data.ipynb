## Flowprint

Remove incorrect folder

```python
models_folder = os.path.join(root_path, 'flowprint/models')
model_names = os.listdir(models_folder)

# remove incorrect model_i folder
for model_name in model_names:
  path = os.path.join(models_folder, model_name, 'data')
  if os.path.exists(path):
    if len(os.listdir(path)) != 101:
      # remove tree of model_i folder
      shutil.rmtree(path)
  else:
    # remove empty model_i folder
    os.rmdir(os.path.join(models_folder, model_name))
```

Get remaining models

```python
models_folder = os.path.join(root_path, 'flowprint/models')
model_names = os.listdir(models_folder)

model_indexes = [int(x.split('_')[1]) - 1 for x in model_names]

remain_indexes = list(set(range(1137)) - set(model_indexes))
len(remain_indexes)
```

Generate data for Flowprint models

```python
train_folder = os.path.join(root_path, 'train')
flowprint_models_folder = os.path.join(root_path, 'flowprint/models')
```

```python
files_dic = dict()
for app in os.listdir(train_folder):
  app_folder = os.path.join(train_folder, app)
  files_dic[app] = os.listdir(app_folder)

# generate data for models
for i in remain_indexes:
  print(i)
  model_folder = os.path.join(flowprint_models_folder, 'model_{}'.format(i+1))
  data_folder = os.path.join(flowprint_models_folder, 'model_{}'.format(i+1), 'data')
  os.mkdir(model_folder)
  os.mkdir(data_folder)

  for app in files_dic.keys():
    filename = files_dic[app][i]
    path1 =  os.path.join(train_folder, app, filename)
    path2 =  os.path.join(data_folder, filename)
    shutil.copy(path1, path2)
```
