## Create sources dataset

Rename the source files

```python
"""
Copy files from raw_data/app/extracted into sources/app and rename a files with 
a more precise form (ex. baohay24h_9_50.056m.csv)
"""

raw_folder = os.path.join(root_path, 'raw_data')
sources_folder = os.path.join(root_path, 'sources')

for app in os.listdir(raw_folder):
  id = 0
  # create app folder
  app_folder1 = os.path.join(raw_folder, app, 'extracted')
  app_folder2 = os.path.join(sources_folder, app)
  
  if not os.path.exists(app_folder2):
    os.mkdir(app_folder2)
    print('Processing {} ...'.format(app))

    for filename in os.listdir(app_folder1):
      id += 1
      print(id)
      file_path = os.path.join(app_folder1, filename)
      df = pd.read_csv(file_path)
      t1 = df['time'].iloc[0]
      t2 = df['time'].iloc[-1]
      t = round((t2 - t1)/60, 3)
      
      new_filename = app + '_' + str(id) + '_' + str(t) + 'm' + '.csv'
      new_path = os.path.join(app_folder2, new_filename)
      shutil.copy(file_path, new_path)
  
```
